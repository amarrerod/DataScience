# Matplotlib

## Importando Matplotlib

Por convenio, matplotlib.pyplot as plt

```python
import matplotlib.pyplot as plt
import numpy as np
```

 ## Definimos el estilo del gráfico

```python
plt.style.use('classic')
```

## Ejemplo básico

```python
x = np.linspace(0, 10, 100)
figure = plt.figure() # Para guardar hay que definir la figura antes que los ejes
plt.plot(x, np.sin(x), '--', label = "Sin(x)")
plt.plot(x, np.cos(x), '.', label = "Cos(x)")
plt.legend()
figure.savefig('mi_figura.png')
plt.show()
```

# Usando subplots

```python
# Primero creamos una rejilla de gráficos
figure, axis = plt.subplots(2)
axis[0].plot(x, 5 * np.sin(x), color = 'k', label = "5Sin(x)")
axis[0].legend()
axis[1].plot(x, np.cos(x), color = 'r', label = "Cos(x)")
axis[1].legend()
figure.savefig('subplots.png')
```
